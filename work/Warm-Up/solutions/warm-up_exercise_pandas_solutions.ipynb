## Select the following columns from the dataset: 

"cname", "wdi_pop", "wdi_popgr", "wdi_gdpcapcur", "wdi_gdpcapgr", "wdi_area", "wdi_broadb", "ht_region"

```
df = data [["cname", "wdi_pop", "wdi_popgr", "wdi_gdpcapcur", "wdi_gdpcapgr", "wdi_area", "wdi_broadb", "ht_region"]]
df
```

## Rename these columns to: "country", "population","population_growth", "gdp_per_capita", "gdp_growth", "area", "internet", "region"

```
df2 = df.rename (columns={"cname" : "name",
            "wdi_pop" : "population",
            "wdi_popgr" : "population_growth",
            "wdi_gdpcapcur" : "gdp_per_capita",
            "wdi_gdpcapgr" : "gdp_growth",
            "wdi_area" : "area",
            "wdi_broadb" : "internet",
            "ht_region" : "region"
           })
df2
```

```
## or, less verbose:
df.columns = ["country", "population","population_growth", "gdp_per_capita", "gdp_growth", "area", "internet", "region"]
df
```

## Create a categorical column from the region codes
the codes (in order correspond to the regions as follows):
"Eastern Europe", "Latin America", "North Africa & Middle East", "Sub-Saharan Africa", "Western Europe and North America", "East Asia","South-East Asia", "South Asia", "Pacific", "Caribbean"

```
regions_dtype = pd.CategoricalDtype([
    "Eastern Europe", "Latin America", "North Africa & Middle East",
    "Sub-Saharan Africa", "Western Europe and North America",
    "East Asia", "South-East Asia", "South Asia", "Pacific", "Caribbean"
])

region_col = pd.Categorical.from_codes(codes=df["region"] - 1, dtype=regions_dtype)

df.loc[:, "region"] = region_col
```

## Select the five countries with the highest population

## Create a new column "population_density"

```
df["population_density"] = pd.to_numeric(df["population"] / df["area"], errors="coerce")
df
```

### What is the per country mean population density for each region
(= sum of densities for each country / number of countries)

```
df["population_density"] = pd.to_numeric(df["population_density"], errors="coerce")
df = df.dropna(subset=["population_density"])
mean_density = df.groupby("region")["population_density"].mean()
```
